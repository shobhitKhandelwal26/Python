# how do I apply a function to a panda series or dataframe

```python
import pandas as pd
```

```python
train=pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

MAP- It is a series method and a dataframe method

APPLY map as a series method

```python
train['Sex_num']=train['Sex'].map({'female':0,'male':1})
```

```python
train.loc[0:4,['Sex','Sex_num']]
```

APPLY- It is both a series method and dataframe method
applies a function to each element in the series

APPLY AS A SERIES METHOD

```python
train['Name_length']=train['Name'].apply(len)
```

```python
train.loc[0:4,['Name','Name_length']]
```

```python
import numpy as np
```

```python
train['Fare_ceil']=train['Fare'].apply(np.ceil)
```

```python
train.loc[0:4,['Fare','Fare_ceil']]
```

```python
train['Name'].str.split(',').head()
```

```python
def get_element(my_list,position):
    return my_list[position]
```

```python
train['Name'].str.split(',').apply(get_element,position=0).head()
```

```python
train['Name'].str.split(',').apply(lambda x:x[0]).head()
```

APPLY AS A DATAFRAME METHOD
applies a function along either axis of dataframe

```python
drinks=pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

```python
drinks.loc[:,'beer_servings':'wine_servings']
```

```python
drinks.loc[:,'beer_servings':'wine_servings'].apply(max,axis=0)
```

```python
drinks.loc[:,'beer_servings':'wine_servings'].apply(max,axis=1)
```

TO identify which column is maximum we use np.argmax function

```python
drinks.loc[:,'beer_servings':'wine_servings'].apply(np.argmax,axis=1)
```

MAP AS A DATAFRAME METHOD
applies to every element of the dataframe

```python
drinks.loc[:,'beer_servings':'wine_servings'].applymap(float)
```

```python
drinks.loc[:,'beer_servings':'wine_servings']=drinks.loc[:,'beer_servings':'wine_servings'].applymap(float)
drinks.head()
```

# Finding all instances of a tag at once

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup.find_all('p')
```

Note that find_all returns a list, so well have to loop through, or use list indexing, it to extract text:

```python
soup.find_all('p')[0].get_text()
```

If you instead only want to find the first instance of a tag, you can use the find method, which will return 
a single BeautifulSoup object:

```python
soup.find('p')
```

# Searching for tags by class and id

```python
page= requests.get('http://dataquestio.github.io/web-scraping-pages/ids_and_classes.html')
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

we can use the find_all method to search for items by class or by id. In the below example, well search 
for any p tag that has the class outer-text:

```python
soup.find_all('p', class_='outer-text')
```

In the below example, well look for any tag that has the class outer-text:

```python
soup.find_all( class_='outer-text')
```

We can also search for elements by id:

```python
soup.find_all(id="first")
```

BeautifulSoup objects support searching a page via CSS selectors using the select method. We can use CSS 
selectors to find all the p tags in our page that are inside of a div like this:

```python
soup.select('div p')
```

Note that the select method above returns a list of BeautifulSoup objects, just like find and find_all.

```python
page=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.XpwiJlMzZQI')
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
seven_day = soup.find(id="seven-day-forecast")
seven_day
```

```python
forecast_items = seven_day.find_all(class_="tombstone-container")
forecast_items
```

```python
tonight = forecast_items[0]
tonight
```

```python
print(tonight.prettify())
```

```python
period = tonight.find(class_="period-name").get_text()
short_desc = tonight.find(class_="short-desc").get_text()
temp = tonight.find(class_="temp").get_text()
print(period)
print(short_desc)
print(temp)
```

```python
img = tonight.find("img")
desc = img['title']
print(desc)
```

# Extracting all the information from the page

```python
period_tags = seven_day.select(".tombstone-container .period-name")
period_tags
```

```python
periods = [pt.get_text() for pt in period_tags]
periods
```

```python
short_descs = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]
temps = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]
descs = [d["title"] for d in seven_day.select(".tombstone-container img")]
print(short_descs)
print(temps)
print(descs)
```

```python
import pandas as pd
weather = pd.DataFrame({
    "period": periods,
    "short_desc": short_descs,
    "temp": temps,
    "desc":descs
})
weather
```

```python
temp_nums = weather["temp"].str.extract("(?P<temp_num>\d+)", expand=False)
weather["temp_num"] = temp_nums.astype('int')
temp_nums
```

```python
weather["temp_num"].mean()
```

```python
is_night = weather["temp"].str.contains("Low")
weather["is_night"] = is_night
is_night
```

```python
weather[is_night]
```
